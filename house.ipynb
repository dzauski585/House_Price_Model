```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

train_df = pd.read_csv("train.csv")
train_df.head(3)
```

```python
train_df.columns
```

```python
train_df = train_df.drop('Id', axis =1 )
```

```python
train_df.info()
```

```python
sns.displot(train_df['SalePrice'], color='g', bins=50)

```

```python
df_num = train_df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
df_num_corr = df_num.corr()['SalePrice']
golden = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print(f"There is {len(golden)} strongly correlated values with SalePrice:\n{golden}")
```

```python
plt.figure(figsize=(9, 7))
sns.heatmap(data=df_num.corr(), vmin=-1, vmax=1, cmap='Spectral').set_title("Correlation Plot")
```

```python
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num, x_vars=df_num.columns[i:i+5], y_vars=['SalePrice'])
```
